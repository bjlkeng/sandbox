```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = pd.concat([pd.Series(np.random.normal(-15, 2., 500)), pd.Series(np.random.normal(5, 1., 1000))])
s
```

```python
proposal = pd.Series(np.random.normal(-15, 2.8, 1000))
proposal

proposal2 = pd.Series(np.random.normal(-15, 10, 1000))
proposal2
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = ax1.twinx()
s.plot.kde(color='black', ax=ax1, linewidth=4, label='Target')

proposal.plot.kde(ax=ax2, linestyle='--', linewidth=2, label='Proposal A')
proposal2.plot.kde(ax=ax2, linestyle='--', linewidth=2, label='Proposal B')
ax1.set_xticks([])
ax1.set_ylabel('')
ax1.set_yticks([])
ax2.set_yticks([])
ax2.set_ylabel('')


ax1.text(-50.5, 0.08, 'Proposal A\nunlikely to\nfind other\nmode', fontsize=15)
ax1.text(-66.5, 0.02, 'Proposal B low\nacceptance rate', fontsize=15)

#ax1.plot(-15.1, 0.045, 'r+', markersize=22)

ax1.legend(fontsize=15)
ax2.legend(loc=(2), fontsize=15)
```
